2) What are the types of machine learning? Describe each with one real-world example.
-  The three primary types of machine learning are supervised learning, unsupervised learning, and reinforcement learning.

i) Supervised Learning
In supervised learning, the algorithm is trained on a labeled dataset, meaning the input data is paired with the correct output. The model learns to predict the output based on the patterns it identifies in the provided input-output pairs. This is akin to a student learning under the guidance of a teacher.

ii) Unsupervised Learning
Unsupervised learning involves training the algorithm on an unlabeled dataset. The algorithm must find structure, patterns, and relationships within the data on its own without guidance.

iii) Reinforcement Learning Reinforcement learning is a goal-oriented approach where an agent learns to make decisions through trial and error within a dynamic environment. The agent receives rewards for desired actions and penalties for undesirable ones, guiding it toward an optimal strategy or policy.

3) Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.

#Overfitting
- Overfitting occurs when a machine learning model learns the details and noise in the training data to the extent that it negatively impacts its performance on new data [1]. The model becomes too complex and essentially "memorizes" the training set, including random fluctuations rather than the true underlying relationship between input and output variables.
Characteristics: High accuracy on the training data but poor performance (generalization) on test or real-world data [1].
Causes: Models that are too complex for the given data (e.g., a deep neural network for a simple linear problem), or insufficient training data relative to the model's complexity [1].
Mitigation: Techniques such as cross-validation, regularization (L1 or L2), reducing model complexity, and using more training data [1].

#Underfitting
Underfitting occurs when a machine learning model is too simple to capture the underlying relationships between input and output variables in the data [1]. The model fails to learn the significant patterns in the training data and thus performs poorly on both training and new data.
Characteristics: Poor accuracy on the training data and even worse performance on the test data [1].
Causes: Using a model that is too simple for the complexity of the data (e.g., a linear regression model for non-linear data), insufficient training time, or using a dataset with too much noise [1].
Mitigation: Using a more complex model, engineering more relevant features, or reducing noise in the data [1].

4) What are outliers in a dataset, and list three common techniques for handling them.

- Outliers are data points that are significantly different or far removed from other observations in a dataset, often indicating errors or unusual occurrences, and they can skew analysis, so common handling techniques include Removal, Capping/Winsorization (replacing extremes with set values), or Transformation (like log transformation) to reduce their impact while retaining the data point.

5) Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data.

- Handling missing values involves identifying them and choosing strategies like deletion or imputation (filling them in) based on data type and missingness patterns, with common methods being Mean/Median imputation for numerical data (e.g., filling with average) and Mode imputation for categorical data (e.g., filling with the most frequent category). Other techniques include advanced methods like KNN or regression for better estimates, or treating missing as a new category for categorical data.

```python
#6) Creates a synthetic imbalanced dataset with make classification() from sklearn.datasets.


from collections import Counter
from sklearn.datasets import make_classification
import pandas as pd

# 1. Create a synthetic imbalanced dataset
# We specify the 'weights' to control the class imbalance (e.g., [0.9, 0.1] for a 90:10 split)
X, y = make_classification(
    n_samples=1000,
    n_features=20,
    n_classes=2,
    weights=[0.9, 0.1],  # Approximately 90% in class 0, 10% in class 1
    random_state=42,     # for reproducibility
    flip_y=0             # set to 0 to prevent label noise from changing the intended distribution
)

# 2. Print the class distribution
print("Original class distribution:", Counter(y))

# Optional: Convert to a pandas DataFrame to see percentage distribution
df = pd.DataFrame({'target': y})
print("\nPercentage distribution:")
print(df['target'].value_counts(normalize=True))
```

```python
#7) Implement one-hot encoding using pandas for the following list of colors: ['Red', 'Green', 'Blue', 'Green', 'Red']. Print the resulting dataframe.

import pandas as pd

# Given list of colors
colors = ['Red', 'Green', 'Blue', 'Green', 'Red']

# Create a DataFrame
df = pd.DataFrame({'Color': colors})

# Apply one-hot encoding
one_hot_df = pd.get_dummies(df, columns=['Color'])

# Print the resulting DataFrame
print(one_hot_df)
```

```python

"""8) Write a Python script to:

- Generate 1000 samples from a normal distribution.

- Introduce 50 random missing values.

- Fill missing values with the column mean.

- Plot a histogram before and after imputation."""


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Generate 1000 samples from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Convert to DataFrame
df = pd.DataFrame({'values': data})

# 2. Introduce 50 random missing values
missing_indices = np.random.choice(df.index, size=50, replace=False)
df.loc[missing_indices, 'values'] = np.nan

# Keep a copy before imputation
df_before = df.copy()

# 3. Fill missing values with the column mean
df['values'] = df['values'].fillna(df['values'].mean())

# 4. Plot histograms before and after imputation
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df_before['values'], bins=30, color='skyblue', edgecolor='black')
plt.title('Before Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(df['values'], bins=30, color='lightgreen', edgecolor='black')
plt.title('After Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
